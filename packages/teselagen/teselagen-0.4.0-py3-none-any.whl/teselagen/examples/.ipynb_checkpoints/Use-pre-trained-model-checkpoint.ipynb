# Use pre-trained model

This notebook shows how to use a pre-trained model in your DISCOVER database to make inference over new data.

```python
import json
import pandas as pd

from teselagen.api.discover_client import DISCOVERClient
from teselagen.api.client import post
```

```python
client = DISCOVERClient(host_url='https://pr-app-7135.teselagen.net/')
```

```python
client.login()
```

```python
#client.get_models_by_type(model_type="predictive")
```

## First Run

```python
data = [
    {'seq':'SCNCVCGVCCSCSP',       'x1': 0, 'x2': 2 , 'y': -38.595},
    {'seq':'STNCFCYPCCSCSS',       'x1': 2, 'x2': 3 , 'y': -25.836},
    {'seq':'RCTCTTIISSSSTF',       'x1': 9, 'x2': 3 , 'y': 10.166},
    {'seq':'LMCTHPLDCSN',          'x1': 1, 'x2': 15, 'y': -3.082},
    {'seq':'KSLRPRCWIKIKFRCKSLKF', 'x1': 3, 'x2': 12, 'y': 31.881},
    {'seq':'RPRCWIKIKFRCKSLKF',    'x1': 0, 'x2': 2 , 'y': 22.94},
    {'seq':'AALKGCWTKSIPPKPCFGKR', 'x1': 2, 'x2': 3 , 'y': 13.973},
    {'seq':'PQRGEGGRAGNLLREEQEI',  'x1': 9, 'x2': 3 , 'y': 44.697},
    {'seq':'SRSELIVHQR',           'x1': 1, 'x2': 15, 'y': 5.388},
    {'seq':'PVVDTTGNNPLQQQEEYYV',  'x1': 3, 'x2': 12, 'y': 11.924},
    {'seq':'ILGKLLSTAAGLLSNL',     'x1': 0, 'x2': 2 , 'y': -18.415},
    {'seq':'FIGKLISAASGLLSHL',     'x1': 2, 'x2': 3 , 'y': -19.259},
    {'seq':'ILGAIIPLVSGLLSHL',     'x1': 9, 'x2': 3 , 'y': -9.165},
    {'seq':'RSNKGFNFMVDMIQALSK',   'x1': 1, 'x2': 15, 'y': 17.069},
    {'seq':'FLPILINLIHKGLL',       'x1': 3, 'x2': 12, 'y': 13.194},
    {'seq':'NFLGTLINLAKKIM',       'x1': 0, 'x2': 2 , 'y': 18.585},
    {'seq':'VLPLISMALGKLL',        'x1': 2, 'x2': 3 , 'y': -13.761},
    {'seq':'FLPIVGKLLSGLL',        'x1': 9, 'x2': 3 , 'y': 3.248},
    {'seq':'GVGDLIRKAVSVIKNIV',    'x1': 1, 'x2': 15, 'y': 14.321},
    {'seq':'GVIDAAKKVVNVLKNLP',    'x1': 3, 'x2': 12, 'y': 10.045}
]
schema_conf = [
    {'id':1,'name': 'seq','value_type': 'aa-sequence', 'type': 'descriptor'},
    {'id':2,'name': 'x1', 'value_type': 'numeric', 'type': 'descriptor'},
    {'id':4,'name': 'x2', 'value_type': 'numeric', 'type': 'descriptor'},
    {'id':3,'name': 'y', 'value_type': 'numeric', 'type': 'target'},
    
]
```

```python
model_job_info = client.submit_model(
    data_input = data,
    data_schema = schema_conf,
    model_type = 'predictive',
    name="Modelo predictivo con secuencias")
```

Now we wait until process finishes on the platform. Then:

```python
# once finished we collect results
data_out = client.get_model_datapoints(
    model_id=model_job_info['modelId'],
    datapoint_type = 'output',
    batch_size = 1000,
    batch_number = 0
)
data_out = pd.DataFrame([el['datapoint'] for el in data_out['data'] ])
display(data_out)
```

## Reusing the model

```python
new_data = [
    {'seq':'SCNCVCGVCCSCSP',       'x1': 7, 'x2': 2},
    {'seq':'STNCFCYPCCSCSS',       'x1': 15.4, 'x2': 3},
    {'seq':'RCTCTTIISSSSTF',       'x1': 19, 'x2': 3.2}
]
```

```python
def submit_pretrained(
    data_input,
    data_schema,
    model_type,
    predictive_modelId,
    configs=None,
    name="",
    description=None):
    body = {
        "dataInput": data_input,
        "dataSchema": data_schema,
        "modelType": model_type,
        "configs": {} if configs is None else configs,
        "name": name,
        "description": "" if description is None else description,
        "predictiveModelId": int(predictive_modelId)
    }
    print(f"body:{body}" )
    print(f"headers: {client.headers}")
    response = post(url=client.submit_model_url,
                    headers=client.headers,
                    json=body)
    response["content"] = json.loads(response["content"])
    return client._get_data_from_content(response["content"])

results = submit_pretrained(
    data_input=new_data,
    data_schema=schema_conf,
    model_type='predictive',
    name='Test with pretrained',
    predictive_modelId=model_job_info['modelId'])
```

```python
# once finished we collect results
data_pre_out = client.get_model_datapoints(
    model_id=results['modelId'],
    datapoint_type = 'output',
    batch_size = 1000,
    batch_number = 0
)
data_pre_out = pd.DataFrame([el['datapoint'] for el in data_pre_out['data'] ])
display(data_pre_out)
```
